##    Determine el orden de la mejor aproximación para las siguientes funciones, usando la Serie de Taylor y el Polinomio de Lagrange:
##   
##     $\frac{1}{25x^2 + 1}, x_0 = 0$
##      $\arctan(x), x_0 = 1$
 
##    Escriba las fórmulas de los diferentes polinomios.
##    Grafique las diferentes aproximaciones.

#### realizado por : CORREA A. FRANCISCO A.

```python
import numpy as np
import matplotlib.pyplot as plt

# 1)
x = np.linspace(-2, 2, 400)

# Función original
f1 = lambda x: 1 / (25 * x**2 + 1)

# Aproximación de Taylor (orden 4)
f1_taylor = lambda x: 1 - 25 * x**2 + (25**2 * x**4) / 2
print("Serie de Taylor para 1 / (25x^2 + 1) en x0 = 0: 1 - 25 * x^2 + (25^2 * x^4) / 2")

# Puntos para el polinomio de Lagrange
x_points = [-1, 0, 1]
y_points = [f1(val) for val in x_points]

# Polinomio de Lagrange usando numpy
lagrange_poly_f1 = np.polyfit(x_points, y_points, 2)
lagrange_f1 = np.poly1d(lagrange_poly_f1)
print("Polinomio de Lagrange para 1 / (25x^2 + 1):", lagrange_f1)

# Gráfica de las  aproximaciones
plt.figure(figsize=(10, 6))
plt.plot(x, f1(x), label='1 / (25x^2 + 1)', color='blue')
plt.plot(x, f1_taylor(x), label='Serie de Taylor (orden 4)', linestyle='--', color='red')
plt.plot(x, lagrange_f1(x), label='Polinomio de Lagrange', linestyle='-.', color='green')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Aproximaciones para 1 / (25x^2 + 1)')
plt.legend()
plt.grid()
plt.show()

# 2)
f2 = lambda x: np.arctan(x)

# Aproximación de Taylor (orden 4) alrededor de x0 = 1
f2_taylor = lambda x: (np.pi/4) + (1/(1 + 1**2)) * (x - 1) - (1/(1 + 1**2)**2) * (x - 1)**2 / 2 + (2*(1/(1 + 1**2)**3)) * (x - 1)**3 / 6
print("Serie de Taylor para arctan(x) en x0 = 1: (pi/4) + (1/(1 + 1^2)) * (x - 1) - (1/(1 + 1^2)^2) * (x - 1)^2 / 2 + (2*(1/(1 + 1^2)^3)) * (x - 1)^3 / 6")

# Puntos para el polinomio de Lagrange
x_points_f2 = [0.5, 1, 1.5]
y_points_f2 = [f2(val) for val in x_points_f2]

# Polinomio de Lagrange usando numpy
lagrange_poly_f2 = np.polyfit(x_points_f2, y_points_f2, 2)
lagrange_f2 = np.poly1d(lagrange_poly_f2)
print("Polinomio de Lagrange para arctan(x):", lagrange_f2)

# Gráfica de las aproximaciones
plt.figure(figsize=(10, 6))
plt.plot(x, f2(x), label='arctan(x)', color='blue')
plt.plot(x, f2_taylor(x), label='Serie de Taylor (orden 4)', linestyle='--', color='red')
plt.plot(x, lagrange_f2(x), label='Polinomio de Lagrange', linestyle='-.', color='green')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Aproximaciones para arctan(x)')
plt.legend()
plt.grid()
plt.show()
```
